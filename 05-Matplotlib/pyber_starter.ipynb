## Total Fares by City Type

```python
# Calculate Type Percents

city_types = combined_city_and_ride_df.groupby("type")

total_fares_by_city_type = city_types.sum()
total_fares_by_city_type.head()

fares_by_type = total_fares_by_city_type.keys()

# Build Pie Chart
fares_pie = total_fares_by_city_type.plot(kind="pie", y=fares_by_type, title="Total Fares by City Type")
fares_pie.set_ylabel("Fares")

```

```python
plt.show()
plt.tight_layout()
plt.axis("equal")


# Save Figure
plt.savefig("../Images/FaresPies.png")
```

## Total Rides by City Type

```python
# Calculate Ride Percents
# Split up our data into groups based upon 'city type'
city_type_group_ride = combined_city_and_ride_df.groupby("type")

total_number_of_rides_per_city = city_type_group_ride["ride_id"].count()
total_number_of_rides_per_city.head()

rides_by_type = total_number_of_rides_per_city.keys()


# Build Pie Chart

rides_pie = total_number_of_rides_per_city.plot(kind="pie", y=rides_by_type, title="Total Rides by City Type")
rides_pie.set_ylabel("Rides")


```

## Total Drivers by City Type

```python
# Calculate Driver Percents
city_types_drivers = combined_city_and_ride_df.groupby("type")

total_number_of_drivers_per_city = city_types_drivers["driver_count"].count()
total_number_of_drivers_per_city.head()

drivers_by_type = total_number_of_drivers_per_city.keys()

# Build Pie Charts
drivers_pie = total_number_of_drivers_per_city.plot(kind="pie", y=drivers_by_type, title="Total Drivers by City Type")
drivers_pie.set_ylabel("Drivers")

```
